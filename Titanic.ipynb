Missing values should be handled for :

Total -1309

age - 1046

fare - 1308

cabin - 295

embarked - 1307

```python
age_mean=round(df['age'].mean(),1)
age_mean
```

```python
fare_mean=round(df['fare'].mean(),1)
fare_mean
```

number - mean

alphabetic - mode

mode ---> return type series, take the 1st series data [0] and fill the missing valuse using fillna()

```python
emb_mode=df['embarked'].mode()[0]
emb_mode
```

Fill the missing data

```python
df['age']=df['age'].fillna(age_mean)
```

```python
df['fare']= df['fare'].fillna(fare_mean)
```

```python
df['embarked']= df['embarked'].fillna(emb_mode)
```

Missing data is handled

```python
df.info()
```

```python
df=df.dropna(axis=1)
```

```python
df.info()
```

                                                Data Transformation                                                              

```python
df=df.rename(columns={'sex':'gender'})
df.columns
```

```python
df['sibsp']+df['parch']
df['family_count']=df['sibsp']+df['parch']
df
```

drop sibsp and parch permanently

```python
df=df.drop(columns=['sibsp','parch'])

df.columns
```

```python
df['embarked'].unique()
```

```python
df['embarked']=df['embarked'].replace({'S':'Southampton','C': 'Cherbourg','Q': 'Queenstown'})
df
```

```python
df.head()
```

Slpitting name as tilte, first name, last name

```python
names= df['name'].str.split(',',expand=True)
names.head()
```

```python
df['last_name']=names[0]
df[['title','first_name']]=names[1].str.split('.',n=1,expand=True) #expand creates separate columns
df
```

After splitting drop the 'name' column as it is unneccessary for the dataset

```python
df=df.drop(columns=['name'])
df
```

```python
for i in ['age','fare']:
  df[i]=round(df[i],2)

df.head()
  
```

Using a cut() method, Create a age group category

```python
df['age_group']=pd.cut(x=df['age'],bins=[0,3,16,25,60,80],labels=['Baby','Child','Teens','Adult','Senior'])
```

Shuffling the columns

```python
df.columns[[9,10,8,1,2,11,5,7,0,3,4,6]]
```

```python
df=df[df.columns[[9,10,8,1,2,11,5,7,0,3,4,6]]]
df.head()
```

Since the fare can't be 0 we are handing the dataset again.  
we don't know that the 0 fare passengers are actual passengers or crew_members. 

so we are considering the 0 fare passengers as crew_memebers and
find the index of the members and remove them permanently from the data set, 
then give reset_index as we have removed some index from the middele of the dataset.

```python
crew_members=df.query('fare==0')

crew_members.index
```

```python
df=df.drop(index=crew_members.index)    

df.reset_index(inplace=True)
df
```
